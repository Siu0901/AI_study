```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
book_data = pd.read_csv('Books.csv')
book_data.head(10)
```

```python
book_data= book_data[book_data['description'] != 'No description available']
book_data.head(50)
```

```python
# 'title' 컬럼을 기준으로 중복된 행을 제거.
# unique 예는 리스트로 반환해서 안됨
book_data = book_data.drop_duplicates(subset=['title'])
book_data.head(50)
```

```python
book_overview = book_data['description']
book_overview[0]
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(book_overview)
tfidf_matrix.shape
```

```python
cosine = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
book_title = dict(zip(book_data['title'], book_data.index))
book_title['Mestizos Come Home!']
```

```python
def get_recommendations(book, top=11):
  idx = book_title[book]
  book_list = cosine[idx]
  book_idx = sorted(list(enumerate(book_list)), key=lambda x:x[1], reverse = True)
  book_idx = book_idx[1:top]
  books_idx = [i[0] for i in book_idx]
  book_recommendation = book_data['title'][books_idx]
  return book_recommendation

books = input()
get_recommendations(books)
```
