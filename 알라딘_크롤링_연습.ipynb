```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = 1  # 원하는 페이지 번호
url = f'https://aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page={page}&cnt=1000&SortOrder=1'

req = requests.get(url)
content = req.content
soup = BeautifulSoup(content, 'html.parser')
```

```python
book_names = soup.select('a.bo3')
book_names[0].text
book_prices = soup.select('span.ss_p2')
int(book_prices[0].text.replace(',','').replace('원', ''))
```

```python
# one page parsing
book_names_parse = [name.text for name in book_names]
book_price_parse = [int(price.text.replace(',','').replace('원','')) for price in book_prices]

# make dataframe & file export
book_df = pd.DataFrame([book_names_parse, book_price_parse], index=['name', 'price']).T
book_df.to_csv('./result.csv', index=False)
book_df
```
