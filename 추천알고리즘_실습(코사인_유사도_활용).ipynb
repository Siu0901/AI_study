```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('movies_metadata.csv',low_memory=False)
data.head()
```

```python
data = data.head(20000)
```

```python
print("결값수: ",data['overview'].isnull().sum())
```

```python
data['overview'] = data['overview'].fillna('')
```

```python
# tfidf = TfidfVectorizer(stop_words='english')
# tfidf_matrix = tfidf.fit_transform(data['overview'])

tfidf = TfidfVectorizer(stop_words='english').fit(data['overview'])
tfidf_matrix = tfidf.transform(data['overview'])
tfidf_matrix.shape

print(tfidf_matrix)
print(tfidf_matrix[1])
# 이게 보면 coords 쪽에 0, 24361 이런식으로 되어있는데
# 0번 영화의 24361번째 단어의 값이 저 values 숫자라는 거임
# 만약 해당 단어가 포함 안되면 values 값이 0임
# 0.135124515 이건 0번 책에 저 단어가 등장했고 중요도가 저정도란 거임 (단어 등장 빈도)
# tf: 해당 문서에서 특정 단어 얼마나 자주 나왔는가 / idf: 전체에서 얼마나 희귀한가
# fit_transform() 하면 단어 리스트 추출하고 각 문서 벡터(숫자 배열로 변환시킴)
# 그리고 실행하면 coords 쪽에 0번 문서 여러개 있잖아 저거 전체가 하나의 행임
# 데이터 타입은 희소행렬이라는데 일단 배열도 아니고 데이터프레임도 아님
# 보통 TF-IDF 행렬은 대부분의 값이 0이기 때문에 다 저장하면 메모리 낭비 심해서
# 0아닌 값들만, 값의 인덱스, 행의 시작 위치만 저장함
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
# 여기부분 ㅈㄴ 햇갈린다
cosine_sim.shape
cosine_sim
```

```python
title_to_index = dict(zip(data['title'], data.index))
idx = title_to_index['Father of the Bride Part II']
idx
```

```python
movie = input("영화 제목을 입력하세요: ")

idx = title_to_index[movie]

sim_scores = list(enumerate(cosine_sim[idx]))

sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

sim_scores = sim_scores[1:11]

movie_indices = [i[0] for i in sim_scores]

result = data.title.iloc[movie_indices]
result
```
