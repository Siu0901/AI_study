word2vec와 메커니즘은 비슷하지만 FastText는 **하나의 단어 안에도 여러 단어들이 존재**하는 것으로 간주한다.

# 1. 내부 단어의 학습
FastText에서 각 단어는 **n**을 몇으로 결정하는지에 따라 **단어들이 얼마나 분리**되는지 결정함.


예를 들어 n = 3이면 apple이란 단어를
**app, ppl, ple**로 분리하고 이들을 백터로 만듦.
여기서 시작과 끝을 의미하는 단어인 <,>를 도입해 **5개의 내부 단어를 백터**로 만듦

<ap, app, ppl, ple, le>
여기서 추가적으로 apple 단어 자체를 백터화함. 결국 단어 n=3 일 때 apple에 대해서

<ap, app, ppl, ple, le>, <apple>
6개로 백터화하는 것

# 2. 모르는 단어(OOV)에 대한 대응
아까처럼 한 단어를 저렇게 세부적으로 나누면 뭐가 좋냐 하는데

가령 birthplace(출생지)란 단어를 학습하지 않은 상태일 때, 다른 단어에 birth와 place란 내부 단어가 있으면 FastText는 birthplace의 벡터를 얻을 수 있음

```python
from gensim.models import FastText

model = FastText(vector_size=100, window=5, min_count=5, workers=4, sg=1)
```
